```python
import matplotlib.pyplot as plt

rankings = {
    2005: ['Superfruit', 'Tea', 'Berry', 'Orange', 'Apple', 'Lemon', 'Strawberry', 'Grape', 'Raspberry', 'Mango'],
    2006: ['Tea', 'Superfruit', 'Berry', 'Orange', 'Mango', 'Grape', 'Strawberry', 'Coffee', 'Unflavored', 'Apple'],
    2007: ['Tea', 'Superfruit', 'Berry', 'Orange', 'Mango', 'Apple', 'Lemon', 'Raspberry', 'Coconut', 'Peach'],
    2008: ['Tea', 'Superfruit', 'Berry', 'Orange', 'Mango', 'Apple', 'Coconut', 'Strawberry', 'Lemonade', 'Lemon'],
    2009: ['Tea', 'Berry', 'Superfruit', 'Orange', 'Apple', 'Coconut', 'Lemon', 'Strawberry', 'Lemonade', 'Coffee']
}

years = list(rankings.keys())
flavors = rankings[2005]  # Assuming flavors remain the same across years

num_cols = 3  # Number of columns for subplots
num_rows = (len(years) + num_cols - 1) // num_cols  # Number of rows for subplots

fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 10))

for i, year in enumerate(years):
    row = i // num_cols
    col = i % num_cols
    ax = axes[row, col] if num_rows > 1 else axes[col]
    ax.barh(flavors, range(1, 11), color='skyblue')
    ax.set_xlabel('Rankings')
    ax.set_title(f'Top 10 Emerging Flavors in {year}')
    ax.invert_yaxis()  # Invert y-axis to display highest ranking at the top

plt.tight_layout()
plt.show()
```

```python
def ranking_flavor(flavor):

    years = [2005, 2006, 2007, 2008, 2009]

    rank = []

    for i in years:
        rank.append(list(rankings[str(i)][rankings[str(i)] == flavor].index))
        
    return rank
```
